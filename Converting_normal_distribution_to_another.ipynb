```python
###Sample code for converting one distribution to another.
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def convert_to_std_distribution(old_dist):
  return (old_dist-old_dist.mean())/old_dist.std()

def convert_from_std_distribution(std_dist, new_mean, new_std):
  return std_dist*new_std+new_mean
```

```python
def convert_distribution(old_dist, new_mean, new_std):
  std_dist= convert_to_std_distribution(old_dist)
  new_dist= convert_from_std_distribution(std_dist, new_mean, new_std)
  return new_dist
```

```python
old_mean=10
old_std=5
size=1000
new_mean=50
new_std=25
```

```python
norm_dist=norm.rvs(loc=old_mean, scale=old_std, size=size)
plt.hist(norm_dist)
plt.show()
```

```python
converted_norm_dist= convert_distribution(norm_dist,new_mean, new_std)
plt.hist(converted_norm_dist)
plt.show()
```
